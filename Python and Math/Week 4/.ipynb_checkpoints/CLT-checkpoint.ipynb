```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

# Chi-square distribution with parameter k

```python
#here you can set k
k = 5

#generate sample of size 1000
chi = sts.chi2(k)
sample = chi.rvs(1000)

#plot histogram and theoretical pdf
x = np.arange(0,4*k,0.1)
y = chi.pdf(x)

plt.hist(sample, normed = True)
plt.plot(x,y)
plt.title("$Chi2(%s)$" %k)
plt.xlabel("$x$")
plt.ylabel("$pdf(x)$")
plt.show()
```

## Central Limit Theorem

```python
# j is the size of sample drawn from the distribution
for j in [5.0, 10.0, 50.0, 100.0]:
    sample_of_means = np.zeros(1000)  #array with sample means

    for i in xrange(len(sample_of_means)): 
        random_sample = np.random.choice(sample,int(j)) #draw sample of size j
        sample_of_means[i] = np.sum(random_sample)/j #count its mean

    #theoretical values for chi2(k):
    #mean = k , var = 2*k
    x = np.arange(k-7, k+7, 0.1)
    norm = sts.norm(loc = k,scale = np.sqrt(2*k/j))  #CLT  
    pdf = norm.pdf(x)

    plt.hist(sample_of_means,normed = True)
    plt.plot(x,pdf)
    plt.title("Sample size: %s" %j)
    plt.xlabel("x")
    plt.ylabel("pdf(x)")
    plt.show()
    
```

### We see that with higher size of sample (j), the approximation with normal distribution is better and better
