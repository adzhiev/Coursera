# Diamonds

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
import pandas as pd
```

```python
data = pd.read_csv('diamonds.txt',sep='\t')
```

```python
X = data.drop('price',axis=1)
y = data['price']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled,y,test_size=0.25,shuffle=True,random_state=1)
```

```python
linear = LinearRegression()
linear.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error
print "Mean squared error for linear model: %f" %mean_squared_error(linear.predict(X_test),y_test)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.scatter(y_test,linear.predict(X_test),color='r');
plt.plot(range(min(y_test),max(y_test),100),range(min(y_test),max(y_test),100),color='b');
```

```python
rand_forest = RandomForestRegressor(random_state=1)
rand_forest.fit(X_train,y_train)
```

```python
print "Mean squared error for random forest model: %f" %mean_squared_error(rand_forest.predict(X_test),y_test)
```

```python
plt.scatter(y_test,rand_forest.predict(X_test),color='r');
plt.plot(range(min(y_test),max(y_test),100),range(min(y_test),max(y_test),100),color='b');
```
